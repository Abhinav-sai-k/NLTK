```python
import gensim
from gensim.models import Word2Vec, keyedvectors
```

```python
from gensim import downloader as api

w2v = api.load("word2vec-google-news-300")
```

```python
w2v['Abhi']
```

```python
w2v.most_similar("Abhi")
```

```python
w2v.most_similar("Telugu")
```

```python
w2v.most_similar("Tamil")
```

```python
# We can also use this model to find how close two words are to each other

print(w2v.similarity("Telugu", "Tamil"))
print(w2v.similarity("Telugu", "Hindi"))
print(w2v.similarity("love","hate"))
print(w2v.similarity("love","ishq"))
```
