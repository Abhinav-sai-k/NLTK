```python
df = pd.read_csv(r"spam.csv",encoding='latin-1')[['v1','v2']]
df.columns = ('y','x')
df
```

### Basic imports

```python
from nltk.stem import SnowballStemmer,WordNetLemmatizer as wnl
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
import re
```

```python
stemmer = SnowballStemmer(language='english')
  
```

```python
len(df)
```

```python
# Storing all the cleaned corpus to a text file with spacing new lines b/w rows
with open('spam_cleaned_corpus.txt','w+') as ptr:

    for i in range(len(df)):
        sens = df['x'][i]
        print(sens)
        sens = re.sub('[^a-zA-Z]',' ',sens)   #Replacing all non alphabetical characters with empty space
        print(sens)
        
        # We perform stemming and stopwords check for the documents
        stemmed_lst = []
        for words in sens.split():  #split by space
            if words not in stopwords.words('english'):
                stemmed_lst.append(stemmer.stem(words))   #Snowball stemming also lowers all the words!
        # print(stemmed_lst)
        ptr.write(' '.join(stemmed_lst)+'\n')
        print()
ptr.close()
            
            
        
```
