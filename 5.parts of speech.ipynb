## This ipynb contains code for finding Parts of speech using nltk.post_tags(words) function

```python
with open('genos.txt','r') as ptr:
    corpus = ptr.read()
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem import WordNetLemmatizer as wnl
lemmatizer= wnl()
from nltk.corpus import stopwords
from nltk import pos_tag 
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tag(['Apple','fear','roar','lover','viral','seed','sow'],lang='eng')
```

```python

for x1,docs in enumerate(sent_tokenize(corpus)):
    lst =[]
    stop_words = []
    words_tokens = word_tokenize(docs)
    for x2,words in enumerate(words_tokens):
        # print(x2,words)
        if lemmatizer.lemmatize(words) not in stopwords.words('english'):
            lst.append(words)
        else:
            stop_words.append(words)
    print(f'Document No : {x1}')
    print("*"*50)
    # print(f"Original, len = {len(words_tokens)} : \n{docs}")
    print(f"After stopwords preprocessing,len = {len(lst)} :\n{' '.join(lst)}")
    #pos tagger code after preprocessing
    print(pos_tag(lst))
    # print(f"stop_words deleted from document, {len(stop_words)} : \n{stop_words}")
    print("*"*50)
    print()
            
        
    print()
```
