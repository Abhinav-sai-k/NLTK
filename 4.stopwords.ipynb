### Apply stopwords and filter the corpus/documents and then apply stemming


```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer(language='english')


for x1,docs in enumerate(sent_tokenize(corpus)):
    lst =[]
    stop_words = []
    words_tokens = word_tokenize(docs)
    for x2,words in enumerate(words_tokens):
        print(x2,words)
        if stemmer.stem(words) not in stopwords.words('english'):
            lst.append(words)
        else:
            stop_words.append(words)
    print(f'Document No : {x1}')
    print("*"*50)
    print(f"Original, len = {len(words_tokens)} : \n{docs}")
    print(f"After stopwords preprocessing,len = {len(lst)} :\n{' '.join(lst)}")
    print(f"stop_words deleted from document, {len(stop_words)} : \n{stop_words}")
    print("*"*50)
    print()
            
        
    print()
```

```python
### Apply stopwords and filter the corpus/documents and then apply Lemmatization
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem import WordNetLemmatizer as wnl
lemmatizer= wnl()


for x1,docs in enumerate(sent_tokenize(corpus)):
    lst =[]
    stop_words = []
    words_tokens = word_tokenize(docs)
    for x2,words in enumerate(words_tokens):
        print(x2,words)
        if lemmatizer.lemmatize(words) not in stopwords.words('english'):
            lst.append(words)
        else:
            stop_words.append(words)
    print(f'Document No : {x1}')
    print("*"*50)
    print(f"Original, len = {len(words_tokens)} : \n{docs}")
    print(f"After stopwords preprocessing,len = {len(lst)} :\n{' '.join(lst)}")
    print(f"stop_words deleted from document, {len(stop_words)} : \n{stop_words}")
    print("*"*50)
    print()
            
        
    print()
```
